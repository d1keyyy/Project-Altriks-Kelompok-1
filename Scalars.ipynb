# SOAL NOMOR 2

```python
import tensorflow as tf
```

```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow
x_tf = tf.Variable(67)
x_tf
```

```python
print (f'tipe data dari x_tf adalah {type(x_tf)}')
```

```python
y_tf = tf.Variable(69)
y_tf
```

```python
print (f'tipe data dari y_tf adalah {type(y_tf)}')
```

```python
sum_tf = x_tf + y_tf
sum_tf = tf.Variable(sum_tf)
sum_tf
```

```python
x_float_tf = tf.Variable(55.0)
x_float_tf
```

```python
sum2_tf = x_float_tf + tf.cast(y_tf, tf.float32)
sum2_tf = tf.Variable(sum2_tf)
print(sum2_tf)
```

```python
print (f'tipe data dari x_float_tf adalah {type(x_float_tf)}')
```
