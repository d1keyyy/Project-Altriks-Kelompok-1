```python
import numpy as np
import matplotlib.pyplot as plt

# data x
x = np.linspace(-10, 10, 1000)

# persamaan
y1 = 2 * x + 1
y2 = 3 * x - 5

# plot
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# batas (opsional, biar fokus ke titik potong)
ax.set_xlim([0, 10])
ax.set_ylim([0, 20])

# gambar garis
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')

# titik potong (6, 13)
plt.axvline(x=6, color='purple', linestyle='--')
plt.axhline(y=13, color='purple', linestyle='--')

# grid & tampilkan
plt.grid()
plt.show()
```
