### SOAL NOMOR 2

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
X_t = X.T
X_t
```

```python
# tampilkan ukuran dari vector x_t
len(X_t)
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
Y = np.array ([55, 7, 4])
Y_t = Y.T
Y_t
```

```python
# tampilkan ukuran dari vector y_t
len(Y_t)
```

### SOAL NOMOR 3

```python
import tensorflow as tf
import torch
```

```python
# lakukan soal nomor 2 menggunakan pytorch dan tensorflow
X_tf = tf.transpose(X)
X_tf
```

```python
X_tf.shape
```

```python
Y_tf = tf.transpose(Y)
Y_tf
```

```python
Y_tf.shape
```

```python
X_pt = torch.transpose(torch.from_numpy(X), 0, 1)
X_pt
```

```python
X_pt.shape
```

```python
Y_pt = torch.transpose(torch.from_numpy(Y.reshape(1, -1)), 0, 1)
Y_pt
```

```python
Y_pt.shape
```
