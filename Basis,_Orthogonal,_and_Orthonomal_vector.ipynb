### SOAL NOMOR 1

```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
I = np.array ([1,0])
I
```

```python
J = np.array ([0,1])
J
```

```python
# buktikan bahwa vector i dan vector j itu orthogonal
np.dot (I, J)
```

### SOAL NOMOR 2

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
V = np.array ([2, 1])
V
```

```python
W = np.array ([2, -1])
W
```

```python
np.dot (V, W)
# V dan W bukanlah vektor yang orthogonal
```
