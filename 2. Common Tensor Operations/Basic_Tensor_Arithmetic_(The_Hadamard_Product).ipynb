```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
import numpy as np
X = np.array([[25, 2, 3], [5, 26, 7]])
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
import torch
X_pt = torch.tensor([[25, 2, 3], [5, 26, 7]])
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
import tensorflow as tf
X_tf = tf.Variable([[25, 2, 3], [5, 26, 7]])
```

```python
# lakukan perkalian matrix X dengan suatu skalar
X*5
```

```python
# lakukan penjumlahan matrix X dengan suatu skalar
X+5
```

```python
# lakukan pengurangan matrix X dengan suatu skalar
X-5
```

```python
# lakukan perkalian matrix X_pt dengan suatu skalar
X_pt*5
```

```python
# lakukan penjumlahan matrix X_pt dengan suatu skalar
X_pt+5
```

```python
# lakukan pengurangan matrix X_pt dengan suatu skalar
X_pt-5
```

```python
# lakukan perkalian matrix X_tf dengan suatu skalar
X_tf*5
```

```python
# lakukan penjumlahan matrix X_tf dengan suatu skalar
X_tf+5
```

```python
# lakukan pengurangan matrix X_tf dengan suatu skalar
X_tf-5
```

```python
# buat matrix Y dengan ukuran 2x3 menggunakan numpy
import numpy as np
Y = np.array([[3, 2, 25], [7, 26, 5]])
```

```python
# buat matrix Y_pt dengan ukuran 2x3 menggunakan pytorch
import torch
Y_pt = torch.tensor([[3, 2, 25], [7, 26, 5]])
```

```python
# buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow
import tensorflow as tf
Y_tf = tf.Variable([[3, 2, 25], [7, 26, 5]])
```

```python
# kalikan matrix X dengan Y
X*Y
```

```python
# kalikan matrix X_pt dengan Y_pt
X_pt*Y_pt
```

```python
# kalikan matrix X_tf dengan Y_tf
X_tf*Y_tf
```
