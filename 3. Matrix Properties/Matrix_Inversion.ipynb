```python
import numpy as np
import tensorflow as tf
import torch
```

### SOAL NOMOR 1

```python
# buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[67,69],[21,41]])
X
```

```python
# buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
Xinv
```

```python
# kalikan X dengan Xinv menggunakan numpy
np.dot(X,Xinv)
```

image.png

```python
# selesaikan persamaan diatas menggunakan numpy
A = np.array([
    [4, 2],
    [-5, -3]
])
y = np.array([4, -7])

hasil_np = np.linalg.solve(A, y)
print(hasil_np)
```

```python
# selesaikan persamaan diatas menggunakan tensorflow
A_tf = tf.constant([
    [4.0, 2.0],
    [-5.0, -3.0]
])
y_tf = tf.constant([
    [4.0],
    [-7.0]
])

hasil_tf = tf.linalg.solve(A_tf, y_tf)
print(hasil_tf)
```

```python
# selesaikan persamaan diatas menggunakan pytorch
A_pt = torch.tensor([
    [4.0, 2.0],
    [-5.0, -3.0]
])
y_pt = torch.tensor([
    [4.0],
    [-7.0]
])

hasil_pt = torch.linalg.solve(A_pt, y_pt)
print(hasil_pt)
```
