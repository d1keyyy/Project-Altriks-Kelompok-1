### SOAL NOMOR 1

```python
# 1. buatlah matriks simetris dengan ukuran 3x3 dan buktikan bahwa itu merupakan matriks simetris

A = np.array([
    [1, 2, 3],
    [2, 4, 5],
    [3, 5, 6]
])

A_transpose = A.T

if np.array_equal(A, A.T):
    print("\nA adalah matriks simetris, karena A = A^T")
else:
    print("\nA bukan matriks simetris")
```

```python
# menggunakan pytorch, buatlah matriks identitas I dengan ukuran 4x4
I = torch.eye(4)
I
```

```python
# menggunakan pytorch, buatlah vector b_pt dengan ukuran 4x1
b_pt = torch.rand(4, 1)
b_pt
```

```python
# kalikan matrix I dengan vector b_pt
hasil = torch.matmul(I, b_pt)
hasil
```
